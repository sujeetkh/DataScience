```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
revenue_train=pd.read_csv('data/patient_monthwise_revenue_train.csv')
#vectorizer = CountVectorizer()
#vectorizer.fit(text)
#vectorizer.vocabulary_
```

```python
revenue_train['diagnosis']=revenue_train['diagnosis'].astype('string')
revenue_train['word_count'] = revenue_train['diagnosis'].apply(lambda x: len(str(x).split(" ")))
revenue_train[['diagnosis','word_count']].head()
```

```python
revenue_train['char_count'] = revenue_train['diagnosis'].str.len()
revenue_train[['diagnosis','char_count']].head()
```

```python
def avg_word(sentence):
    try:
        words = sentence.split()
    except:
        return 1
    return (sum(len(word) for word in words)/len(words))

revenue_train['avg_word'] = revenue_train['diagnosis'].apply(lambda x: avg_word(x))
revenue_train[['diagnosis','avg_word']].head()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
revenue_train['stopwords'] = revenue_train['diagnosis'].apply(lambda x: len([x for x in x.split() if x in stop]))
revenue_train[['diagnosis','stopwords']].head()
```

```python
revenue_train['numerics'] = revenue_train['diagnosis'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
revenue_train[['diagnosis','numerics']].head()
```

```python
revenue_train['upper'] = revenue_train['diagnosis'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
revenue_train[['diagnosis','upper']].head()
```

```python
revenue_train['diagnosis'] = revenue_train['diagnosis'].apply(lambda x: " ".join(x.lower() for x in x.split()))
revenue_train['diagnosis'].head()
```

```python
def last_word_digit(sent):
    if sent=='':
        return None     
    vals=sent.split()
    if vals[len(vals)-1].isdigit():
        return vals[len(vals)-1] 
    return None

val=tmp['diagnosis'].apply(lambda x:last_word_digit(str(x))) 
```

```python
val.value_counts()
```

```python
def diag_val(diag,nps,un):
    val=str(diag)
    if not nps.isdigit() and not nps=='nan':
        val=val+' '+ nps
    if not un.isdigit() and not un=='nan':
        val=val+' '+un
    return val
        
tmp['diag']=tmp.apply(lambda row:diag_val(row['diagnosis'],str(row['avg_nps']),str(row['Unnamed: 20'])), axis=1)
```

```python
tmp['diag']=tmp['diag'].str.replace('[^\w\s]','')
```

```python
tmp['diag']=tmp['diag'].str.replace('_','')
```

```python
from textblob import TextBlob
tmp['diag']=tmp['diag'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
def nps_val(nps,un):
    val=str(nps)
    if val.isdigit():
        return val
    if un.isdigit():
        return un  
    return None
        
tmp['nps']=tmp.apply(lambda row:nps_val(str(row['avg_nps']),str(row['Unnamed: 20'])), axis=1)
```
